````python
#자판기 프로그램
def beverages():
    coke = 600
    cider = 800
    water = 1000

    money = int(input("금액 넣기 >>"))
    menu = int(input("번호 입력 >>"))

    if menu == 1:
        if money - coke >= 0:
            remine = money - coke
            thousands = remine // 1000
            fhundreds = (remine - 1000*thousands) // 500
            ohundreds = (remine - 1000*thousands - 500*fhundreds) // 100
            print(f"거스름돈 1000원 {thousands}개 500원 {fhundreds}개 100원 {ohundreds}개")
        else:
            print("금액이 부족합니다.")
            
    elif menu == 2:
        if money - cider >= 0:
            remine = money - cider
            thousands = remine // 1000
            fhundreds = (remine - 1000*thousands) // 500
            ohundreds = (remine - 1000*thousands - 500*fhundreds) // 100
            print(f"거스름돈 1000원 {thousands}개 500원 {fhundreds}개 100원 {ohundreds}개")
        else:
            print("금액이 부족합니다.")

    elif menu == 3:
        if money - water >= 0:
            remine = money - water
            thousands = remine // 1000
            fhundreds = (remine - 1000*thousands) // 500
            ohundreds = (remine - 1000*thousands - 500*fhundreds) // 100
            print(f"거스름돈 1000원 {thousands}개 500원 {fhundreds}개 100원 {ohundreds}개")
        else:
            print("금액이 부족합니다.")

    else:
        print("메뉴가 없습니다.")
        beverages()
````

````python
score_dict = {'이름' : ['주미리', '신재영', '김호성'], 
              '파이썬' : [85, 90, 80], 
              '기초통계' : [100, 70, 90]}

#과목 관련 변수명
subjects_name = []
subjects = 0

#합계 관련 변수명
score_merge = []
scores = 0
mean_merge = []
mean = 0

#반복문으로 score_dict을 돌면서 과목 수, 합계를 누적계산하기
for i in score_dict:
    if i != '이름':
        subjects_name.append(i)
        subjects += 1
    else:
        subjects = subjects

for j in range(0, len(score_dict['이름'])):
    cal_scores = 0
    for i in subjects_name:
        scores = score_dict[i][j]
        cal_scores += scores
    score_merge.append(cal_scores)

#반복문으로 구한 합계/과목수 로 평균 구하기
for i in score_merge:
    mean = i / subjects
    mean_merge.append(mean)

for j in range(0, len(score_dict['이름'])):
    print(f"{score_dict['이름'][j]} 학생의 전체 과목 평균은 {mean_merge[j]}점 입니다.")
````

````python
# 키오스크 만들기
# 1. 사용자에게 주문할 메뉴 이름을 입력받는다
# 2. 그 메뉴가 목록에 있으면 가격을 안내한다
#    삼겹살 -> 삼겹살[12000] 주문되었습니다
# 3. 만약 없는 메뉴를 입력하면 "없는 메뉴입니다" 안내
# 4. 사용자가 "quit"을 입력하면 주문을 종료하고, 
# 5. 주문 총 금액을 계산해서 출력하기
#    총 주문 금액 000원 입니다.

def keyorder():
    menu = {
        "삼겹살" : 12000,
        "목살" : 13000,
        "소주" : 5000,
        "맥주" : 4500,
        "공깃밥" : 1000,
    }
    
    total = 0
    order_list = []
    
    while True:
        user = input("주문할 메뉴 >> ")
        if user in menu:
            cost = menu[user]
            total += cost
            print(f"{user} 메뉴는 {cost}원 주문되었습니다.")
        elif user == "quit":
            total = total
            break
        else:
            print("없는 메뉴입니다.")
            break
            
    print(f"총 주문 금액은 {total}원 입니다.")
    
keyorder()
````
