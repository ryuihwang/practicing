- 목표 : 특성 4개를 활용해서 붓꽃의 품종을 예측하는 모델

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

from sklearn.datasets import load_iris
```

2. 데이터 수집
- 학습용으로 제작된 iris 데이터를 로드

```python
#번치 객체 : {} 번치 객체, 파이썬에서 딕셔너리와 굉장히 유사, 머신러닝 데이터셋
#data : 문제 데이터
#target : 정답 데이터
#target_names : 정답 데이터 이름
#feature_names : 컬럼명
#DESCR : 데이터셋에 대한 설명

iris_data = load_iris()
iris_data
```

```python
#문제데이터와 정답데이터의 형태
print(iris_data['data'].shape)
print(iris_data['target'].shape)
```

```python
iris_df = pd.DataFrame(iris_data['data'], columns = iris_data['feature_names'])
iris_df
```

```python
#문제와 정답 선언
X = iris_df
y = iris_data['target']
```

```python
#학습, 평가 데이터로 최종 분리 (***랜덤 샘플링***)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=8) #테스트 데이터의 비율)
```

```python
X_train
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# 산점도 매트릭스로 시각화 진행해보기
# x축과 y축에 특성을 넣고 여러개의 산점도를 한번에 출력
# 여러 특성쌍 관계를 동시에 확인 가능
# 특성의 개수가 너무 많으면 확인하기가 힘들어짐
```

5. 모델링
   - 모델선택, 학습, 평가

```python
knn_model = KNeighborsClassifier(n_neighbors=10)
```

```python
#학습
knn_model.fit(X_train, y_train)
```

```python
pred = knn_model.predict(X_test)
```

```python
#모델이 예측한 결과값과 실제 정답을 넣고 정확도 평가
metrics.accuracy_score(pred, y_test)
```

```python
knn_model.score(X_test, y_test)
```

```python
knn_model.score(X_train, y_train)
```

```python
#정확도 외 다른 지표들
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
my_info = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]],
                       columns=iris_df.columns)

predict = knn_model.predict(my_info)

print("예측 결과:", predict)
print("종 이름:", iris_data['target_names'][predict])
```
