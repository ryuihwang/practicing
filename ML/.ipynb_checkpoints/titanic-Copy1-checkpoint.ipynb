```python
#AgeëŠ” Pclassì™€ ìƒê´€ê´€ê³„ê°€ ìƒëŒ€ì ìœ¼ë¡œ ë†’ì•˜ìŒ
#Pclass, sexë¥¼ êµ¬ë¦…í™”í•˜ì—¬ age í‰ê·  ì¶œë ¥
gb1 = train[['Pclass', 'Sex', 'Age']].groupby(['Pclass', 'Sex']).mean()
gb1
```

```python
#2ë“±ê¸‰ ê°ì‹¤, ë‚¨ì„±ì¼ ê²½ìš°ì˜ í‰ê·  ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
gb1.loc[(2,'male'),:].iloc[0] #ê°’í•˜ë‚˜ë§Œ ë½‘ì•„ì£¼ê¸°
```

```python
#train Age ê²°ì¸¡ì¹˜ í–‰
train[train['Age'].isnull()]
```

```python
#ê·œì¹™ì— ë§žê²Œ Pclassì™€ Sexë¥¼ ë³´ë©´ì„œ Age ì±„ì›Œë„£ëŠ” í•¨ìˆ˜ ì •ì˜
def fill_age(row):
    # row : ë°›ì•„ì˜¤ëŠ” í–‰ì— ë°ì´í„°ì— ëŒ€í•œ ë§¤ê°œë³€ìˆ˜
    # rowì˜ ageê°€ ê²°ì¸¡ì¹˜ë¼ë©´ gb1ì—ì„œ ë§¤ì¹­ë˜ëŠ” ê°’ì„ ì°¾ì•„ì„œ ì¸ë±ì‹± í•´ì˜¤ë„ë¡
    # ê²°ì¸¡ì¹˜ê°€ ì•„ë‹ˆë¼ë©´ ì›ëž˜ ìž‘ì„±ëœ ë‚˜ì´ê°’ì„ ì‚¬ìš©
    
    #ë§Œì•½ì— ë‚˜ì´ê°€ ê²°ì¸¡ì¹˜ë¼ë©´, gb1ì˜ ê°’ì„ ì¸ë±ì‹±
    if np.isnan(row['Age']) : #ë°›ì•„ì˜¨ í–‰ì˜ Ageê°€ ê²°ì¸¡ì¹˜ë¼ë©´
        return gb1.loc[(row['Pclass'],row['Sex']),:].iloc[0]

    else:
        return row['Age']
```

```python
train['Age'] = train.apply(fill_age, axis=1).astype(np.int64)
```

```python
test['Age'] = test.apply(fill_age, axis=1).astype(np.int64)
```
