```python
#알고리즘이 클러스터 중심으로 삼을 데이터 포인트 3개를 무작위로 지정
#할당된 포인트의 평균값으로 클러스터 중심(centroid)를 업데이트 (반복)
#클러스터 중심에 할당되는 데이터 포인트의 변화가 없을때까지 반복
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
data = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
data
```

```python
#y는 비지도학습이기 때문에 별도로 빼놓기
labels = pd.DataFrame(iris.target)

labels.columns = ['labels']
labels
```

```python
feature = data[['sepal length (cm)', 'sepal width (cm)']]
feature.head()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(
    n_clusters = 3, #클러스터 개수(k수) 설정
    algorithm='lloyd',
    #로이드 알고리즘은 기본값
    n_init = 10, #초기 중심점을 무작위로 선택하는 횟수 -> 10번 정도 초기 중심점을 다르게 해서 진행해보고 그중 가장 좋은 결과를 선택
    random_state = 2025
)
```

```python
#학습
model.fit(feature)
```

```python
#예측 어떤 꽃인지 알고 있지 않음. 그냥 군집 0,1,2로 분류한 것
pred = model.predict(feature)
pred
```

```python
#예측한 컬럼을 feature에 합치기
pred_df = pd.DataFrame(pred, columns=['predict'])
```

```python
pred_df
```

```python
df = pd.concat([feature, pred_df], axis=1)
df
```

```python
#교차표 : 두 열의 값들이 각각 몇번 나타나는지 카운팅
# df['predict']와 labels['labels'] 비교

ct = pd.crosstab(labels['labels'], df['predict'])
ct
#해서 ㄱ: 실제값 0인 데이터 포인트를 예측 레이블 0으로 50개 모두 분류했다. 
#라벨의 0과 클러스터 군집 0은 다른 의미
```

```python
#centroid로 확인
model.cluster_centers_
```

```python
#cenetroid 위치
centers = pd.DataFrame(model.cluster_centers_, columns=feature.columns)
```

```python
centers
```

```python
df['predict'].value_counts()
```

```python
#데이터 포인터들을 centroid와 함께 시각화
plt.scatter(
    df['sepal length (cm)'], #x축 좌표 설정
    df['sepal width (cm)'], #y축 좌표 설정
    c = df['predict'], #예측된 군집 레이블에 따라 색상을 다르게
    alpha = 0.5
)

#centroid의 x축 좌표
center_x = centers['sepal length (cm)']
center_y = centers['sepal width (cm)']
#centroid의 시각화
plt.scatter(
    center_x,
    center_y,
    marker = 'D', c = 'r'
)
plt.show()
```

```python
#군집화가 어느 정도 효율적으로 잘 되었는지를 평가할 수 있는 지표들이 별도로 존재함
#지도학습(ex분류)과 비지도학습의 평가지표 다름
#clustering의 경우 같은 클러스터에 같은 포인트가 들어있는지, 다른 클러스터 간에 다른 포인트가 들어가 있는지가 중요
```

```python
#비지도 학습 평가 지표 - ARI
#실제 라벨이 필요함
#대부분의 군집화 데이터세트는 ARI나 AMI처럼 비교할만한 실제 레이블을 가지고 있지 않음.
#RI : 가능한 모든 쌍의 경우에 대해 정답인 쌍의 개수를 비율로 
#ARI : RI의 기댓값을 RI 값에서 빼서, 기댓값과 분산을 재조정 (-1~1)
#ARI : 각 일치하는 쌍의 비율
#AMI : 두 클러스터링에서 겹치는 정보의 비율
```

```python
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
# 예측한 라벨값 확인
model.labels_
#model.predict(feature)
```

```python
#ARI
adjusted_rand_score(labels['labels'], model.labels_)
```

```python
#AMI
adjusted_mutual_info_score(labels['labels'], model.labels_)
```

```python
# 둘 다 0.6 정도 근사치로, 실제값과 어느정도 일치하기는 하나, 완벽하지는 않다고 해석됨
# 실제 레이블 값이 있으면, 지도 학습 모델을 만들기에, 비지도 학습 모델에 적합하다고 볼 수는 없음.
# ARI나 AMI의 경우에는 성능평가보다는, 알고리즘을 개발하거나 특정 프로세스에서 사용된다.
```

```python
#비지도 학습 평가 지표 - 실루엣 분석
# 실루엣 계수 : 클러스터링이 잘되었는지 확인할 수 있는 지표(-1~1)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
#평균 실루엣 확인
silhouette_score(feature, pred)
```

```python
silhouette_vals = silhouette_samples(feature, pred)
silhouette_vals
```

```python
df_sil = pd.concat([df, pd.Series(silhouette_vals, name = 'silhouette_coef')], axis=1)
df_sil
```

```python
#각 클러스너별 실루엣 계수의 평균값
df_sil.groupby('predict')['silhouette_coef'].mean()

#어떠한 레이블 상관없이 실루엣 계수가 낮음
```

```python
#최적의 클러스터 수 결정
# 엘보우 기법
# SSE(클러스터 내 데이터 포인터들이 클러스터 중심으로부터 얼마나 떨어져 있는지를 나타내는 값)가 가장 완만해지는 k의 수를 찾는 것

model.inertia_ #이게 SSE값 : 군집화 한 이후에, 각 중심점에서 데이터 포인트간의 거리의 제곱을 합산
```

```python
#엘보우 그리기
#군집 개수 범위 설정
ks = range(1,10) #군집개수 k
inertias = [] #sse 담아줄 리스트

for k in ks:
    model = KMeans(
        n_clusters=k,
        n_init = 10, #초기 중심점 무작위로 선택하는 횟수
        random_state=2025
    )
    #각각의 군집 개수로 학습
    model.fit(feature)
    inertias.append(model.inertia_)
```

```python
inertias
```

```python
#군집 개수에 따른 시각화
plt.plot(ks, inertias, marker= 'o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertias, SSE')


plt.show()
```

```python
#클러스터의 수가 3일 떄, 그래프의 기울기가 크게 완만해짐을 관측
#k가 3일 때, 급격한 감소가 멈추고, 그 이후로는 더 많은 클러스터를 추가하더라도 SSE가 크게 감소하지 않는다.
# 3이 가장 합리적인 군집 개수라고 볼 수 있다.
```
