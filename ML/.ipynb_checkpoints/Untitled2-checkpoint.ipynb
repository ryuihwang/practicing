###선형회귀 모델 데이터 특성확장
- 선형회귀 모델은 데이터의 분포가 직선의 형태를 벗어나게 되면 정확도가 낮아짐.
- 또한, 선형회귀 모델은 하이퍼파라미터가 존재하지 않음

```python
#copy는 기존의 데이터는 그대로 두면서, 복사하여 새로운 변수명 넣어주기
#만약 copy 없이 새로운 변수에 담아준다면, 내용이 바뀔때 같이 바뀔 수가 있다.
extended_X_train = X_train.copy()
```

```python
extended_X_train
```

```python
#feature extended

for col1 in X_train.columns:
    for col2 in X_train.columns:
        extended_X_train[col1 + 'x' + col2] = X_train[col1] * X_train[col2]

#13x13=169개 열 추가
```

```python
extended_X_train
```

```python
#test도 같은 작업
extended_X_test = X_test.copy()
```

```python
#feature extended

for col1 in X_test.columns:
    for col2 in X_test.columns:
        extended_X_test[col1 + 'x' + col2] = X_test[col1] * X_test[col2]

#13x13=169개 열 추가
```

```python
extended_X_test
```

```python
#특성 확장된 train 데이터로 학습
linear_model2 = LinearRegression()
linear_model2.fit(extended_X_train, y_train)
```

```python
linear_model2.score(extended_X_train, y_train)
```

```python
linear_model2.score(extended_X_test, y_test)
```

```python
#라쏘 모델 적용
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso()
```

```python
lasso_model.fit(extended_X_train, y_train)
```

```python
sum(lasso_model.coef_ != 0)
```

```python
#라쏘는 중요도가 낮은 특성의 w값을 0으로 만들기 때문에 0이 아닌 값들의 개수를 합해서 사용된 특성수를 알 수 있다.
print(f'사용한 특성 수 : {sum(lasso_model.coef_ != 0)}')
#182개중 94개 사용, 88개의 가중치는 0이됨
```

```python
lasso_model.score(extended_X_train, y_train)
```

```python
lasso_model.score(extended_X_test, y_test)
```

- train에서의 r2는 내려갔지만, test에서는 오름
- 선형회귀 모델에 비해 과대적합이 줄어들고 성능 향상

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=1)
```

```python
ridge_model.fit(extended_X_train, y_train)
```

```python
ridge_model.score(extended_X_train, y_train)
```

```python
ridge_model.score(extended_X_test, y_test)
```

- alpha값이 1일 때는 릿지모델에 비해 라쏘모델이 과대적합 해소에는 더 적합하다고 볼 수 있다.

하이퍼파라미터 튜닝
- alpha값을 바꿔가며, RMSE 확인

```python
#lasso 모델
#alpha값에 따른 RMSE 값을 빈 리스트에 담아주자
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
lasso_train_list = []
lasso_test_list = []

for i in alpha_list:
    lasso_model = Lasso(alpha = i)
    lasso_model.fit(extended_X_train, y_train)
    
    #예측값 구하기
    lasso_train_pre = lasso_model.predict(extended_X_train)
    #RMSE
    lasso_train_RMSE = mean_squared_error(lasso_train_pre, y_train)**0.5
    #빈 리스트에 RMSE값 넣기
    lasso_train_list.append(lasso_train_RMSE)

    #예측값 구하기
    lasso_test_pre = lasso_model.predict(extended_X_test)
    #RMSE
    lasso_test_RMSE = mean_squared_error(lasso_test_pre, y_test)**0.5
    #빈 리스트에 RMSE값 넣기
    lasso_test_list.append(lasso_test_RMSE)
```

```python
lasso_train_list
```

```python
lasso_test_list
```

```python
#한글을 쓰기 위한 코드
from matplotlib import rc
rc('font', family = 'AppleGothic')
```

```python
plt.figure(figsize=(10,5))
plt.plot(lasso_train_list, label='train') #train 그래프
plt.plot(lasso_test_list, label='test') #test 그래프
plt.title('Lasso 모델 : alpha - RMSE 그래프') #그래프 제목
plt.xlabel('alpha값')
plt.xticks(np.arange(6), alpha_list)
plt.ylabel('RMSE')
plt.legend()
plt.grid()
plt.show()
```

```python
#라쏘 모델에서는 alpha값이 0.1일때 가장 성능이 좋다.
```

```python
#ridge 모델
#alpha값에 따른 RMSE 값을 빈 리스트에 담아주자
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
ridge_train_list = []
ridge_test_list = []

for i in alpha_list:
    ridge_model = Ridge(alpha = i)
    ridge_model.fit(extended_X_train, y_train)
    
    #예측값 구하기
    ridge_train_pre = ridge_model.predict(extended_X_train)
    #RMSE
    ridge_train_RMSE = mean_squared_error(ridge_train_pre, y_train)**0.5
    #빈 리스트에 RMSE값 넣기
    ridge_train_list.append(ridge_train_RMSE)

    #예측값 구하기
    ridge_test_pre = ridge_model.predict(extended_X_test)
    #RMSE
    ridge_test_RMSE = mean_squared_error(ridge_test_pre, y_test)**0.5
    #빈 리스트에 RMSE값 넣기
    ridge_test_list.append(ridge_test_RMSE)
```

```python
plt.figure(figsize=(10,5))
plt.plot(ridge_train_list, label='train') #train 그래프
plt.plot(ridge_test_list, label='test') #test 그래프
plt.title('ridge 모델 : alpha - RMSE 그래프') #그래프 제목
plt.xlabel('alpha값')
plt.xticks(np.arange(6), alpha_list)
plt.ylabel('RMSE')
plt.legend()
plt.grid()
plt.show()
```
