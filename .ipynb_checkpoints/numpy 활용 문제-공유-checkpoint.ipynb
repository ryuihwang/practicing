#### Q5. 합계 구하기 
 1. 전체 원소의 합 구하기
 2. 각 행(row)별 합 구하기
 3. 각 열(column)별 합 구하기

풀 수 있는 방법 1 - hint 1. 전치 사용하기  
풀 수 있는 방법 2 - hint 1. np.sum()의 인자 확인해보기

```python
rows = arr1.shape[0]
columns = arr1.shape[1]
row_sum = []
column_sum = []

for i in range(0, rows, 1):
    value = np.sum(arr1[i, :])
    row_sum.append(value)

for j in range(0, columns, 1):
    values = np.sum(arr1[:, j])
    column_sum.append(values)

print(row_sum)
print(column_sum)
```

#### Q7. 각 행의 0번 열이 50을 초과하는 경우 그 행이 몇 번째 행인지, 행의 원소들이 무엇인지 출력 

```python
for i in range(0, rows, 1):
    if arr1[i, 0] >= 50:
        print(f"{i}행, {arr1[i, 0]}")
```

#### Q8. 각 행의 두번째 값이 0을 포함한 짝수인 경우 그 행의 3번째 이후의 값을 1이상 20 이하의 수로 변경 후 출력 

```python
lists = []

for i in range(0, rows, 1):
    if arr1[i, 1] % 2 == 0:
        lists = arr1[i, 2:]
        for j in range(0, len(lists), 1):
            if lists[j] > 20:
                lists[j] = np.random.randint(1,20)
        print(f"{i}행, {lists}")
```

#### Q9. 7*7 단위 행렬을 생성(arr1)하고, 행렬을 뒤집어서 저장(arr2) 및 출력
  
아래 힌트보다 여러 종류의 풀이가 가능합니다  
hint 1. index/slicing의 step값을 생각해보기 

```python
eye1 = np.eye(7)
eye2 = np.eye(7)

for i in range(0, eye1.shape[0], 1):
    for j in range(0, eye1.shape[1], 1):
        j2 = 6-j
        i2 = 6-i
        eye2[i,j] = eye1[i2, j2]
        
print(eye1)
print(eye2)
```

#### Q10. eye1의 각 행의 1의 위치가 짝수인 경우 1의 값을 랜덤(100>= x >=200)으로 변경하고, 홀수인 경우 99로 출력(저장 x)
hint 1. np.where(condition) # condition: 조건

```python
eye1_ = eye1

for i in range(0, len(eye1), 1):
    if eye1[i, 1] % 2 == 0:
        eye1_[i, 1] = np.random.randint(100,200)
    else:
        eye1_[i, 1] = 99

print(eye1_)
```

#### Q11. eye1과 eye2의 각 원소값을 'or'연산하여 출력 (boolean 자료형)

```python
result = (eye1 != 0) | (eye2 != 0)
print(result)
```
