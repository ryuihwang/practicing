```python
#2015~2017 광주광역시 범죄현황 데이터
###목적: 데이터를 이용해서 전년 대비 지역별 범죄 증감률 구하기
- 증감률 : (금년발생건수 - 작년발생건수) / 작년 * 100
```

```python
import pandas as pd
```

```python
#데이터 로드
#인덱컬럼은 '관서명'
#인코딩 'euc-kr'
data_2015 = pd.read_csv("data/2015.csv", encoding='euc-kr', index_col = '관서명')
data_2016 = pd.read_csv("data/2016.csv", encoding='euc-kr', index_col = '관서명')
data_2017 = pd.read_csv("data/2017.csv", encoding='euc-kr', index_col = '관서명')
```

```python
data_2015['총합'] = data_2015.iloc[:,1:6].sum(axis=1)
data_2016['총합'] = data_2016.iloc[:,1:6].sum(axis=1)
data_2017['총합'] = data_2017.iloc[:,1:6].sum(axis=1)
```

```python
print(data_2015.shape)
print(data_2016.shape)
print(data_2017.shape)
```

```python
print(set(data_2015.index))
print(set(data_2016.index))
print(set(data_2017.index))
```

```python
###데이터 전처리
#데이터를 살펴보면, df2017에만 광주지방경찰청 인덱스가 있음. 데이터도 부실함. 
#해당 행 삭제(drop함수 통해서)
element = set(data_2017.index) - set(data_2015.index)
```

```python
data_2017.drop(element, inplace=True)
```

```python
print(set(data_2015.index))
print(set(data_2016.index))
print(set(data_2017.index))
```

```python
df_2015_occur = data_2015.loc[:,'총합'][data_2015.iloc[:,0] == '발생건수']
df_2016_occur = data_2016.loc[:,'총합'][data_2016.iloc[:,0] == '발생건수']
df_2017_occur = data_2017.loc[:,'총합'][data_2017.iloc[:,0] == '발생건수']
```

```python
print(df_2015_occur)
print(df_2016_occur)
print(df_2017_occur)
```

```python
print(df_2015_occur.shape)
print(df_2016_occur.shape)
print(df_2017_occur.shape)
```

```python
df_2015_to_2016 = (df_2016_occur - df_2015_occur) / df_2015_occur * 100
```

```python
df_2015_to_2016
```

```python
df_2016_to_2017 = (df_2017_occur - df_2016_occur) / df_2016_occur * 100
```

```python
df_2016_to_2017
```

```python
### 시리즈들을 데이터프레임으로 묶기
total = pd.concat([df_2015_occur, df_2016_occur, df_2017_occur, df_2015_to_2016, df_2016_to_2017], axis=1)
total.columns = ['2015총계', '2016총계','2017총계','2015~2016증감률','2016~2017증감률']
```

```python
total.style.format('{:.2f}') #소수점 둘째자리까지
total
```

```python
total2 = pd.concat([df_2015_occur, df_2016_occur, df_2017_occur], axis=1)
```

```python
total2
```

```python
import matplotlib.pyplot as plt
#x,y 값 설정해서 선그래프 그리기
y_data = total2.iloc[0,:]
x_data = [2015, 2016, 2017]
plt.plot(x_data,y_data, ls='--', lw=5, c='violet', marker='P', mfc='blue', mec='green', ms=13, mew=1.5)
plt.xticks([2015, 2016, 2017])
```
