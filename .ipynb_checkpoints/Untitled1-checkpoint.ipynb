```python
movie = np.genfromtxt('data/ratings.dat', delimiter = '::', dtype=np.int64)
```

```python
print(movie)
```

```python
np.savetxt('data/temp.txt', movie, delimiter=':', fmt='%d')
```

```python
#데이터 속성 확인
#배열 형태
print(movie.shape)
#배열 차원
print(movie.ndim)
#배열 전체 요소 수
print(movie.size)

#전체 영화의 평점들의 평균
np.mean(movie[:, 2])
```

```python
#전체 영화의 평점들의 평균
np.mean(movie[:, 2])
```

```python
#아이디가 1인 사용자가 매긴 영화의 평점들의 평균
np.mean(movie[:, 2][movie[:, 0] == 1])
```

```python
len(movie[movie[:,0]== 2])
```

```python
#데이터 유효성 검사
#중복값 제거
arr = np.unique(movie[:, 0])
arr
```

```python
#각 사용자가 매긴 평점들의 평균
#빈 리스트에 담기
#아이디는 1~6040까지 6040종류의 아이디가 있다고 가정
mean_list = []
mean_value = 0
for i in arr:
    mean_value = np.mean(movie[:, 2][movie[:, 0] == i])
    mean_list.append([i, mean_value])
```

```python
len(mean_list)
```

```python
mean_list
```

```python
#각 사용자별 평균 평점이 4점 이상인 사용자 id만 구해보기
mean_array = np.array(mean_list)
print(mean_array[:, 0][mean_array[:, 1] >= 4].astype(int))
```
