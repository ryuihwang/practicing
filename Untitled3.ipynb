```python
hr_info = pd.read_csv('data/hr-info.csv', encoding='euc-kr')
hr_info.head(5)
hr_info.describe(include=['object'])
hr_info.describe(include=['float'])

#sum
#전체 총 지출 인건비 확인
hr_info['SALARY'].sum()

#mean()
#평균 연봉 확인
hr_info['SALARY'].mean()

#count
#퇴사자 수 확인
#결측치는 계산에 포함 x
hr_info['RETIRE_DATE'].count()

#남/녀 비율
hr_info['GENDER'].value_counts()

#직원들의 국적 종류 알아보기
hr_info['NATION'].unique()

#최대, 최소 근속연수 확인
print("최대:", hr_info['SERVICE_YEAR'].max())
print("최소:", hr_info['SERVICE_YEAR'].min())

#박사 학위 취득자의 평균 연봉
hr_info['SALARY'][hr_info['EDU_LEVEL'] == '박사 학위'].mean()
```

```python
#학사 학위 취득자의 평균 연봉
hr_info['SALARY'][hr_info['EDU_LEVEL'] == '석사 학위'].mean()
```

```python
#교육수준별 평균 연봉(반복문으로)
for i in hr_info['EDU_LEVEL'].unique():
    mean_value = hr_info['SALARY'][hr_info['EDU_LEVEL'] == i].mean()
    print(i, ":", mean_value)
```

```python
hr_info[['EDU_LEVEL','SALARY']].groupby(['EDU_LEVEL']).mean()
```

```python
###그룹화 후 여러 통계값 확인가능
#교육수준과 성별에 따른 평균, 최대, 최소, 중앙 연봉 값 확인
hr_info[['EDU_LEVEL','GENDER','SALARY']].groupby(['EDU_LEVEL','GENDER']).describe()
#agg 함수 사용해서 여러가지 집계함수 동시에 확인 가능
hr_info[['EDU_LEVEL','GENDER','SALARY']].groupby(['EDU_LEVEL','GENDER']).agg(['mean','max','min','median'])
```

```python
###그룹화 후 컬럼별로 다른 연산 적용
#국적에 따른 평균 (평균)성과평가점수와 (최대)연봉 확인
hr_info[['NATION','PERFORMANCE_SCORE','SALARY']].groupby(['NATION']).agg({'PERFORMANCE_SCORE':'mean', 'SALARY':'max'})
```
