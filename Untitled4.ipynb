```python
pay_info = pd.DataFrame()
for i in file_names:
    df = pd.read_csv(i, encoding='euc-kr')
    pay_info = pd.concat([pay_info, df])
```

```python
#org_info에 각 조직별 조직 인원 수, 평균나이, 평균근무년수, 평균연봉 추가
org_info = pd.read_csv('data/org-info.csv', encoding='euc-kr', index_col='ORG_CD')
```

```python
org_info
```

```python
hr_info = pd.read_csv('data/hr-info.csv', encoding='euc-kr')
```

```python
hr_info.groupby('ORG_CD').count()
```

```python
#조직정보데이터랑 인사정보데이터에서 공통넉으로 갖고 있는 컬럼 ORG_CD
#ORG_CD를 기준으로 groupby -> 즉, org_info와 인덱스가 같아짐

#hr_info에서 groupby를 이용해서 부서별 부서 인원수, 평균나이, 평균근무년수, 평균연봉 구하기
or_agg = hr_info.groupby('ORG_CD').agg({'ENAME' : 'count', 'AGE' : 'mean', 'SERVICE_YEAR' : 'mean', 'SALARY' : 'mean'})
```

```python
or_agg['AGE'] = or_agg['AGE'].astype(int)
```

```python
or_agg
```

```python
#org_info와 hr_info 데이터에서 추출한 df을 가로로 병합
final_df = pd.concat([org_info, or_agg], axis=1)
```

```python
final_df
```

```python
#rank의 기본값: 작은 수가 1위(ascending=True)
#큰수가 1위 : (ascending=False)
or_agg['SALARY'].rank(ascending=False)

#or_agg에 "부서별 연봉 순위" 열을 새로 만들어서 데이터를 대입
or_agg['SALARY_RANK'] = or_agg['SALARY'].rank(ascending=False)

#랭크를 기준으로 부서별 연봉 순위가 1위인 순위부터 자세히 살펴보기
or_agg.sort_values('SALARY_RANK')
```
