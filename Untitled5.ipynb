```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###연도별 입사자 추세 확인
#parse_dates : 날짜 타입으로 특정 컬럼 변환
hr_info = pd.read_csv('data/hr-info.csv', encoding='euc-kr', parse_dates=['HIRE_DATE'])
hr_info.head()
```

```python
hr_info.info()
```

```python
#입사연도 추출
hr_info['HIRE_DATE'].dt.year
```

```python
hr_info['입사년도'] = hr_info['HIRE_DATE'].dt.year
```

```python
#연도별로 입사자 수 파악
hr_info['입사년도'].value_counts().sort_index(ascending=True)
```

```python
#연도별로 입사자 수 선그래프로 파악
hr_info['입사년도'].value_counts().sort_index(ascending=True).plot()
```

```python
#한글 폰트 적용
plt.rc('font', family='AppleGothic')
```

```python
#학위별 평균 연봉 비교
hr_info[['EDU_LEVEL', 'SALARY']].groupby(['EDU_LEVEL']).mean().plot(kind='bar', rot=45)
```

```python
#학위 및 성별에 따른 평균연봉 시각화
hr_info[['EDU_LEVEL', 'GENDER', 'SALARY']].groupby(['EDU_LEVEL', 'GENDER']).mean().plot(kind='bar', rot=45)
```

```python
#교육 수준과 성별에 따른 평균 연봉
salary_mean = hr_info[['EDU_LEVEL', 'GENDER', 'SALARY']].groupby(['EDU_LEVEL', 'GENDER']).mean().round(1)
```

```python
salary_mean.reset_index()
```

```python
#데이터 재구조화
salary_mean2 = salary_mean.reset_index().pivot(index='EDU_LEVEL', columns='GENDER', values='SALARY')
```

```python
salary_mean2
```

```python
#수평으로 된 바차트
salary_mean2.plot(kind='barh', colormap='Set1', xlabel='평균연봉', ylabel='교육수준', 
                  title='교육수준과 성별에 따른 평균연봉 비교', subplot=True)
```

```python
#부서별 평균 연봉 비교하기
#조직정보 불러오기
org_info = pd.read_csv("data/org-info.csv", encoding='euc-kr')
org_info.head()
```

```python
hr_info
```

```python
#팀별 평균 연봉 구하는 목적 -> 팀정보는 org_info, 연봉 정보는 hr_info에 있음
#merge
org_hr_merge = pd.merge(hr_info, org_info, on = 'ORG_CD')
hr_info[['ORG_CD', 'SALARY']].groupby(['ORG_CD']).mean()
```

```python
org_hr_merge = pd.merge(hr_info, org_info, on = 'ORG_CD')
```

```python
team_salary_mean = org_hr_merge[['KOR_TEAM', 'SALARY']].groupby(['KOR_TEAM']).mean()
```

```python
team_salary_mean.plot(kind='bar', figsize=(20,5), rot=45, ylim=(4000,7000))
```

| 옵션        | 의미 |
|------------|------|
| values     | - 원 그래프에 들어갈 데이터 |
| labels     | - 원 그래프 바깥에 표시되는 데이터의 Name |
| autopct    | - 소수점 자리표시 설정.<br> - %.1f%% = 소수점 1자리까지 표시하고 뒤에 % 표시 |
| startangle | - 원 그래프의 첫 데이터 시작 위치를 설정<br> - 90 = 원 그래프의 첫 데이터 시작을 12시 기준으로 설정 |
| counterclock | False = 원 그래프 출력 시 시계방향으로<br> True = 원 그래프 출력 시 시계 반대방향으로 |

```python
#각 부서별 인원수 확인
org_hr_merge['KOR_DIVISION'].value_counts()
```

```python
org_hr_merge['KOR_DIVISION'].value_counts().plot(kind='pie',
                                                startangle=90, #90도 방향에서 시작(12시 방향)
                                                counterclock = False, #시계방향으로 가게 됨
                                                autopct = '%.1f%%', #소수점 첫째자리까지 표시해주셈
                                                title = '각 본부의 인원 비율'
                                                )

plt.savefig('data/pie.png')
```

```python
#나이에 따른 연봉 변화
#상관계수가 1에 가깝다고 볼 수 있음
hr_info.plot(kind='scatter', x='AGE', y='SALARY', color='red')
```

```python
import seaborn as sns

#성별에 따른 연봉 히스토그램
sns.histplot(#seaborn의 히스토그램을 생성하는 함수
    hr_info,
    x='SALARY',
    hue='GENDER', #분할 기준
    multiple='stack', #막대를 쌓아서(stacked) 그룹별 분포 확인
    palette = ['#5471CFFF','#CF5479FF'],
    linewidth = 2.0
    
)
```

```python
#버블차트 : 스캐터에서 버블크기가 추가된 정도
#나이와 근속 연수에 따른 부서별 연봉 비교
sns.relplot(data=hr_info,
          x='AGE', y='SERVICE_YEAR', hue='POS_CD', #구분기준은 직위 코드
           size='SALARY', #버블크기 기준, 원이 클수록 연봉 높다
            sizes=(40, 400), #버블크기 적용 사이즈
            #alpha=0.5 #투명도
          )
```

```python
hr_corr = hr_info[['PERFORMANCE_SCORE', 'SALARY', 'AGE', 'SERVICE_YEAR', 'TRAINING_HOURS']].corr()
```

```python
sns.heatmap(hr_corr, annot=True, fmt='.2f', linewidth=0.5)
```
